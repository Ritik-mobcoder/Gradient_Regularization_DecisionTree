```python
from sklearn.datasets import make_regression
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression
```

## Gradient Descent

Gradient Descent is a fundamental optimization algorithm widely used in machine learning and data science to minimize a cost function by iteratively moving in the direction of the steepest descent.

```python
X,y = make_regression(n_samples=100, n_features=1, n_informative=1, n_targets=1,noise=20,random_state=13)
plt.scatter(X,y)
```

```python
# Initialize and fit a Linear Regression model
reg = LinearRegression()
reg.fit(X, y)
```

```python
print("Coefficient:", reg.coef_)
print("Intercept:", reg.intercept_)
```

```python
plt.plot(X, reg.predict(X), color="Red", label="OLS (Ordinary Least Squares)")
plt.legend()
plt.show()
```

```python
m = 60  # Arbitrary initial slope
b = 0
```

```python
loss_slope = -2 * np.sum(y - m * X.ravel() - b)
print("Loss slope:", loss_slope)
```

```python
lr = 0.1
```

```python
step_size = lr * loss_slope
print("Step size:", step_size)
```

```python
b = b - step_size
print("Updated intercept (b):", b)
```

```python
y_pred = (m * X + b).reshape(-1)
```

```python
plt.scatter(X, y, label="Data")
plt.plot(X, reg.predict(X), color="Red", label="OLS (Ordinary Least Squares)")
plt.plot(X, y_pred, color="Green", label="Gradient Descent")
plt.legend()
plt.show()
```
